```python
data['хэштег'].values
```

```python
string = ((data['хэштег']+' ')*data['частотность']).values
```

```python
string
```

```python
import itertools
text = []
for arr in string:
    text.append(arr.split())
text = list(itertools.chain(*text))
```

```python
error_tags = ['кнн']
```

```python
text = [word for word in text if word not in error_tags]
```

```python
text
```

```python
text = [word.replace('_', ' ') if 'сексуальное' not in word else word for word in text ] 
```

```python
to_wordcloud = ' '.join(text)
```

```python
#to_wordcloud = to_wordcloud.replace('_', ' ')
```

```python
import os
from PIL import Image
import wordcloud
from os import path
from wordcloud import WordCloud
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt


figure(figsize=(20, 10), dpi=80)

stopwords = set()

_mask = np.array(Image.open("../диплом/07052022/07052022/m1.png"))
_mask


wordcloud = WordCloud(mask=_mask,
                      width= 3000, height = 2000, stopwords=stopwords,background_color="white" , 
                      contour_width=2, contour_color='white', collocations=False).generate(to_wordcloud)

#plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
wordcloud.to_file("../диплом/07052022/07052022/cloud.png")
plt.axis("off")
plt.show()
```

# Graph Womans

```python
import networkx as nx
from itertools import combinations
G = nx.Graph()
colors_i = ['red', 'blue', 'yellow', 'pink', 'green', 'purple', 'black']
```

```python
import pandas as pd
toks = pd.DataFrame()

for i in range(0,7):
    one_public = pd.read_excel('../женские паблики (токсичные).xlsx', sheet_name=i)
    #print(one_public)
    one_public['public'] = i
    G.add_nodes_from(one_public.Слово.values)
    G.add_edges_from(list(combinations(one_public.Слово.values, 2)), color=colors_i[i])
    toks = toks.append(one_public)
```

```python
one_public
```

```python
toks.Слово.values
```

```python
toks_summ = toks.groupby('Слово').sum('Частотность').reset_index()
```

```python
G.remove_node('кнн')
G.nodes
```

```python
list_sizes = []
for node in G.nodes:
    if node != 'кнн':
        list_sizes.append(toks_summ[toks_summ.Слово == str(node)]['Частотность'].values[0])
```

```python
edges,colors = zip(*nx.get_edge_attributes(G,'color').items())
```

```python
from matplotlib.pyplot import figure, text
from matplotlib.lines import Line2D

figure(figsize=(20, 10), dpi=80)
pos = nx.spring_layout(G)
d = dict(G.degree)
nx.draw_circular(G,with_labels=True,
               node_color='pink', 
               alpha=0.8,
               node_size=list_sizes,font_size=21,edge_color=colors, width=0.8)

handles = [Line2D([], [], color=color, label=label) for color, label in zip(colors_i, ["Куны Не Нужны (КНН)",
                                                                                       "Подслушано Феминизм", 
                                                                                       "Terra Feminarum", 
                                                                                       "Территория либерального матриархата",
                                                                                       "Русский радфем - бессмысленной и беспощадный",
                                                                                       "Рофлим над мужским эго", 
                                                                                       "•Wrong Fem• | Феминизм"])]
ax.legend(handles=handles)
plt.legend(handles=handles)
plt.show()

plt.savefig('../диплом/07052022/07052022/woman_graph.png')
```

# Проценты

```python
with open('../../Частотные_слова_c_леммитизацией_мужскойинформатор.txt') as f:
    data = f.read().split('\n')
```

```python
words = []
numbers = []
for word in data:
    l = word.replace('(', '').replace(')','').replace("'", '').replace("'",'').split(', ')
    try:
        w, n = l[0], l[1]
    except:
        continue
    words.append(w)
    numbers.append(n)
```

```python
numbers = map(int,numbers)
```

```python
resdf = pd.DataFrame({'Слово':words, 'Количество':numbers})
```

```python
resdf['Процент'] = round(resdf.Количество*100/resdf.Количество.sum(),2)
```

```python
resdf.to_excel('Процент_частотных_слов_мужскойинформатор.xlsx')
```
